```python
'''10. How can you tell the difference between break and continue?

Ans: Both 'break' and 'continue' are control flow statements in Python, but they have different effects on the execution of program's.

'break' is used to exit a loop prematurely. When break is encountered in a loop, the loop is immediately terminated, and the program execution continues with the next statement after the loop.

Ex: '''
for i in range (1,10):
    if i == 5:
        break
    print(i)
    
#This above code will print the numbers 1 through 4 and then exit the loop when i is equal to 5, because the break statement is encountered.

'''continue is used to skip the current iteration of a loop and move on to the next iteration. When continue is encountered in a loop, the current iteration is terminated, and the loop continues with the next iteration.

Ex: '''

for i in range(1,10):
    if i % 2 == 0:
        continue
    print(i)
    
#This program will print the odd numbers between 1 and 9 and skip the even numbers, because the continue statement is encountered when i is even.
```

```python
'''11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

Ans: In a for loop, range(10), range(0, 10), and range(0, 10, 1) all produce the same sequence of integers, from 0 to 9 inclusive. However, they differ in the way they specify the start, stop, and step values of the sequence.

range(10) produces a sequence of integers starting from 0 (the default start value) and ending at 9 (stop value specified as 10 - 1). It increments by a step of 1 (the default step value).

Ex: '''

for i in range(10):
    print(i)
    
'''range(0, 10) produces a sequence of integers starting from 0 (specified start value) and ending at 9 (stop value specified as 10 - 1). It increments by a step of 1 (the default step value).

Ex:'''

for i in range(0,10):
    print(i)
    
'''range(0, 10, 1) produces a sequence of integers starting from 0 (specified start value) and ending at 9 (stop value specified as 10 - 1). It increments by a step of 1 (specified step value).

Ex:'''

for i in range(0,10,1):
    print(i)
```

```python
'''12. Write a short program that prints the numbers 1 to 10 using a for loop.
Then write an equivalent program that prints the numbers 1 to 10 using a while loop.'''

#using for loop
for i in range(1,11):
    print(i)
    
#using while loop
i = 1
while i<=10:
    print(i)
    i = i+1
```
